```python
import matplotlib.pyplot as plt

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criar gráfico
plt.plot(x, y)

# Título e rótulos
plt.title("Gráfico Simples")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# Mostrar gráfico
plt.show()
```

```python
###Likes vs Views (dispersão)
plt.figure(figsize=(8,6))

plt.scatter(df["Views"], df["Likes"], alpha=0.5)

plt.title("Views vs Likes")
plt.xlabel("Views")
plt.ylabel("Likes")

plt.show()
```

```python
### Energia das músicas
plt.hist(df["Energy"], bins=20, color="orange")

plt.title("Distribuição de Energia")
plt.xlabel("Energy")
plt.ylabel("Quantidade")

plt.show()
```

```python
###Top 10 Músicas com Mais Streams
import pandas as pd
import matplotlib.pyplot as plt

# Top 10 músicas
top10 = df.sort_values(by="Stream", ascending=False).head(10)

# Gráfico
plt.figure(figsize=(12,6))

plt.barh(top10["Track"], top10["Stream"], color="purple")

plt.title("Top 10 Músicas com Mais Streams")
plt.xlabel("Streams")
plt.ylabel("Música")

plt.gca().invert_yaxis()

plt.show()
```

```python
top3 = df.sort_values(by="Views", ascending=False).head(4)
plt.figure(figsize=(7,5))
plt.barh(top3["Track"], top3["Views"],color="red")
plt.title("Top 3 músicas com mais views")
plt.ylabel("Música")
plt.gca().invert_yaxis()
plt.show()
```
