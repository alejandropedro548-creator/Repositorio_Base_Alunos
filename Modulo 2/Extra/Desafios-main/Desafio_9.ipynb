### Listas

1. Soma dos Elementos

```python
# Dada uma lista de números, calcule a soma de todos os elementos
numeros = [5, 10, 15, 20, 25]
soma = sum(numeros)

# Seu código aqui

print(f"Soma: {soma}")  # Deve imprimir 75
```

2. Encontrar o Maior Número

```python
# Encontre o maior número em uma lista
numeros = [34, 12, 89, 5, 67, 23]
maior = max(numeros)

# Seu código aqui

print(f"Maior número: {maior}")  # Deve imprimir 89
```

3. Adicionar Elementos com Condição

```python
# Dada uma lista de números, crie uma nova lista apenas com números maiores que 50
numeros = [25, 60, 15, 80, 45, 90, 30]
maiores_que_50 = []

# Seu código aqui (use append com if)
for i in range(len(numeros)):
    if numeros[i] > 50:
        maiores_que_50.append(numeros[i])

print(maiores_que_50)  # Deve imprimir: [60, 80, 90]
```

4.  Criar Lista de Vogais de uma Frase

```python
# Crie uma lista com todas as vogais de uma frase
frase = "Python é uma linguagem de programação"
vogais = []

# Seu código aqui (use append com if)
for i in range(len(frase)):
    if frase[i] == "a" or frase[i] == "e" or frase[i] == "i" or frase[i] == "o" or frase[i] == "u" or frase[i] == "é":
        vogais.append(frase[i])
print(vogais)  # Deve imprimir: ['o', 'é', 'u', 'a', 'i', 'u', 'a', 'e', 'o', 'a', 'a', 'o']
```

5. Controle de Gastos Pessoais

```python
print("=== CONTROLE DE GASTOS PESSOAIS ===")

# Listas para armazenar os dados
nomes_gastos = []
valores_gastos = []

# Pergunta a quantidade de gastos
quantidade = int(input("Digite a quantidade de gastos: "))

# Loop para coletar os gastos
for i in range(quantidade):
    nome = input("Digite o nome do produto: ")
    valor = float(input("Digite o valor do produto: "))
    
    # Adiciona às listas usando APPEND
    nomes_gastos.append(nome)
    valores_gastos.append(valor)

# Calcula o total dos gastos
total_gastos = sum(valores_gastos)

# Exibe os resultados
print("\n=== RESUMO DOS GASTOS ===")
for i in range(quantidade):
    print(f"{nomes_gastos[i]} - R$ {valores_gastos[i]:.2f}")

print(f"\nTOTAL GASTO: R$ {total_gastos:.2f}")

# Verifica se está gastando muito
if total_gastos >= 1500:
    print("⚠️ Você está gastando muito!")
else:
    print("✅ Seus gastos estão sob controle.")
```
